```python
#모듈 임포트
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
wines = load_wine() #데이터 변수 할당
wines.keys()
wines_data = wines.data #feature data
wines_label = wines.target #target data
#label data를 feature data 맨 뒤에 붙일 예정임
wines.keys()
#wines_data.shape -> 데이터 크기 확인
```

```python
wines.target #target array
wines.target_names #target names
```

```python
print(wines.DESCR)
```

```python
import pandas as pd #데이터 시각화 해서 확인먼저 하기

wines_df = pd.DataFrame(data=wines_data, columns=wines.feature_names)
wines_df
```

```python
#feature data에 label data 삽입
wines_df["label"] = wines_label #columns 이름 label
wines_df
#마지막 열에 label data가 삽입된 것을 알 수 있다.

```

```python
# (3) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(wines_data, 
                                                    wines_label, 
                                                    test_size=0.2, 
                                                    random_state=50)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test),"\n",'y_train개수: ', len(y_train),"y_test개수: ",len(y_test))
#train_test_split(x value, y value, test_size = ~~, random_state = ~)
#test_size = 0.2 => 20%만 사용한다 테스트에
#random_state = => 얼마나섞을것인지 난수 만들기
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

#model 학습부분
decision_tree = DecisionTreeClassifier(random_state=32) #모델 종류, 난수설정
decision_tree.fit(X_train, y_train) #모델 학습시키기(fitting)
y_pred = decision_tree.predict(X_test) #y_pred에 모델 예측값 삽입


accuracy = accuracy_score(y_test, y_pred) #당연히 크기가 같아야겠쥬?
print("accuracy : ",accuracy) #정답데이터/전테데이터
print(classification_report(y_test, y_pred, target_names = wines.target_names))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

#model 학습부분
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)


accuracy = accuracy_score(y_test, y_pred) #당연히 크기가 같아야겠쥬?
print("accuracy : ",accuracy,"%") #정답데이터/전테데이터
print(classification_report(y_test, y_pred,target_names = wines.target_names))

#?? 정확도가 100%가 나왔다, overfitting인가? spilit시 난수 설정 했는데 그래도?
```
