# 인물사진 01. 정면 셀카 사진 블러처리하기

## 필요 라이브러리 임포트 및 파일 가져오기

```python
import os
import urllib
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt

print('슝=3')
```

```python
img_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/image_3.png'  
img_orig = cv2.imread(img_path) 

print(img_orig.shape)
plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
plt.show()
```

## Segmentation에 사용되는 모델 가져오기
- PixelLib에서 제공하는 세그멘테이션 모델 이용

```python
# 저장할 파일 이름을 결정합니다
model_dir = os.getenv('HOME')+'/aiffel/human_segmentation/models'
model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5')

# PixelLib가 제공하는 모델의 url입니다
model_url = 'https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5'

# 다운로드를 시작합니다
urllib.request.urlretrieve(model_url, model_file)
```

### 세그멘테이션 모델 생성 - model 변수에 할당

```python
model = semantic_segmentation()
model.load_pascalvoc_model(model_file)
```

### 모델에 이미지 입력
- "segmentAsPascalvoc" : PASCAL VOC 데이터로 학습된 모델을 이용한다는 의미

```python
segvalues, output = model.segmentAsPascalvoc(img_path)

# model.segmentAsPascalvoc(img_path)
```

### model에서 나온 출력값

```python
plt.imshow(output)
plt.show()
```

결과가 다른 색상으로 담겨있다

```python
segvalues
```

- class_ids : 0(background), 15(person)이 있는 것을 알 수 있음
- masks : T/F값

### PASCAL VOC 데이터의 라벨 종류

```python
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
len(LABEL_NAMES)
```

### segvalues에 있는 LABEL_NAME 모두 찾기

```python
for class_id in segvalues['class_ids']:
    print(LABEL_NAMES[class_id])
```

### 물체마다 Output 색 파악하기

```python
colormap = np.zeros((256, 3), dtype = int) #빈행렬 만들기 rgb 색생계로 색상 256가지 나타냄
ind = np.arange(256, dtype=int) # 0~255 까지 숫자 생성

for shift in reversed(range(8)):
    for channel in range(3):
        colormap[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3

colormap[:20]
```

```python
#사람 나타내는 색
print("사람 나타내는 색 :",colormap[15])

#배경 나타내는 색
print("배경 나태내는 색 :",colormap[0])
```

### 객체 마스크 만들기

output image가 BGR 순서로 되어있으므로, 사람 나타내는 색의 배치를 반대로 만든다.

(192 128 128) >>>> (128,128,192)

```python
seg_color = (128,128,192)

# output의 픽셀 별로 색상이 seg_color와 같다면 1(True), 다르다면 0(False)
seg_map = np.all(output==seg_color, axis=-1) 


print(seg_map.shape)
plt.imshow(seg_map, cmap='gray')
plt.show()
```

#### 원래 이미지와 겹쳐 보자

```python
img_show = img_orig.copy()

# True과 False인 값을 각각 255과 0으로 바꿔줍니다
img_mask = seg_map.astype(np.uint8) * 255 #1-> 255 / 0 -> 0

# 255와 0을 적당한 색상으로 바꿔봅니다
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)

# 원본 이미지와 마스트를 적당히 합쳐봅니다
# 0.6과 0.4는 두 이미지를 섞는 비율입니다.
img_show = cv2.addWeighted(img_show, 0.4, color_mask, 0.6, 0.0)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("[Background segmentation result]")
plt.show()
```

### 블러처리하기 - blur() 함수 사용

```python
# (13,13)은 blurring kernel size를 뜻합니다
img_orig_blur = cv2.blur(img_orig, (50,50))  #커널 사이즈를 올릴수록 블러 심해짐
plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("[blurred image result]")
plt.show()
```

### 흐려진 이미지에서 세그멘테이션 마스크만 이용해서 배경 추출

```python
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)
img_bg_mask = cv2.bitwise_not(img_mask_color)
img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("[blurred background segmentation result]")
plt.show()
```

### 만든 배경(img_bg_blur)과 원본이미지(img_orig)합치기

```python
img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)
# 세그멘테이션 마스크가 255인 부분만 원본 이미지 값을 가지고 오고 아닌 영역은 블러된 이미지 값을 사용

fig = plt.figure(figsize=(10, 6))
fig.add_subplot(1,2,1)
plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
plt.title("[Original picture]")
plt.axis("off")

fig.add_subplot(1,2,2)
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.title("[After processing]")
plt.axis("off")


plt.show()
```

### Result
- 주변 배경 화면이 잘 블러처리 되었다.

- np.where 함수를 이용하여 세그멘테이션 마스크가 255인 부분, 즉 얼굴 부분에 해당하는 곳에는 원본 이미지를 가져왔고, 아닌 영역은블러된 사진에서 가져왔다.

- 머리카락처럼 미세한 부분은 person으로 인지하지 않고 배경으로 인식해, 잔머리 몇가닥이 사라져 어색하게 되었다.

- person 객체 주변이 그림자처럼 어울지는데, 이는 커널 크기를 작게 조절하면 조금 개선되는것으로 보인다.

## 해결 방법 제안

#### 세그멘테이션 성능 강화하기
세그멘테이션이 잘 되면 여러가지 주요 문제들을 해결할 수 있다. 먼저 3D 카메라를 이용하는 방법이 있다. 차세대 3D 이미지 센서 핵심 기술‘D-ToF’는 직접 ToF로, 물체로 보낸 신호가 수신부에 되돌아오는 시간을 측정해 거리를 유추한다. 이는 수십, 혹은 수백미터 이상 떨어진 물체와의 거리도 측정할 수 있는 장점이 있다. 또 기존 모델과 달리 센서를 소형화할 수 있다는 점에서 실용성이 높다. 이를 이미지 세그멘테이션에 도입하면 이미지 인식률을 더 높일 수 있을 것이다.


##### 현재 우리가 사용했던 pixelib은 pointrend라는 기법을 이용해 정확도를 높인다.
   
   
이는 기존의 R-CNN의 이미지 인식보다 훨씬 훌륭하게 세그멘테이션을 진행하지만, 그럼에도 불구하고 이미지 색상이 겹치거나, 여러 객체가 복잡하게 얽혀있는 경우 오류가 발생하곤 한다.

##### 3D 카메라 기법 D-ToF는 픽셀별 인식이 아닌, 사진에  전압을 걸어 자유 전자의 수를 증폭시키는 '충격 이온화' 현상을 이용한다.

   
    
보다시피 외부 이미지 센서로 조사된 광자로 인해 생섣되는 자유 전자수는 매우 크게 증가하는 것을 볼 수 있다. 미세한 수준의 광자 량으로도 이 광자를 증폭해 아주 많은 광자가 들어온 것처럼 인식할 수 있다. 즉, 어둡거나 복잡한 사진도 정밀히 판별할 수 있게 된다. 위 기술을 이용하면 이미지 세그멘테이션이 더 효과적일 것으로 예상된다.

#### 다른 기술과 융합하기

구글에서는 물체의 온도를 측정하는 적외선 카메라를 함께 이용해 센싱을 한다고 한다. 이처럼 다른 기술과도 융합하여 세그멘테이션 효과를 증대시킬 수도 있다. 개인적으로 집단 지성의 힘을 믿는 사람으로서, 세그멘테이션 분야에서도 블렌딩 기법을 써 보는건 어떨까 싶다. 적외선카메라, 3D 카메라 기법, HOG, R-CNN, Pointrend 등 현존하는 다양한 이미지 인식 기법들을 모두 합쳐 average pooling 등으로 다수의 모델이 선택한 값으로 세그멘테이션을 진행하는 것이다. 각각의 모델마다 더 잘 인식되는 부분이 있을 것이기 때문에 이는 유용하다 생각한다. 또 잔 오류들을 이를 통해 줄여줄 것으로 예상된다.
