```python
#데이터 가져오기
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
cancer = load_breast_cancer() #데이터 변수 할당
cancer.keys()
cancer_data = cancer.data #feature data
cancer_label = cancer.target #target data
#label data를 feature data 맨 뒤에 붙일 예정임
cancer.keys()
#wines_data.shape -> 데이터 크기 확인
```

```python
cancer.target #target array
cancer.target_names #target names
```

```python
print(cancer.DESCR)
```

```python
import pandas as pd #데이터 시각화 해서 확인먼저 하기

cancer_df = pd.DataFrame(data=cancer_data, columns=cancer.feature_names)
cancer_df
```

```python
#feature data에 label data 삽입
cancer_df["label"] = cancer_label #columns 이름 label
cancer_df
#마지막 열에 label data가 삽입된 것을 알 수 있다.
```

```python
# (3) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(cancer_data, 
                                                    cancer_label, 
                                                    test_size=0.3, 
                                                    random_state=7)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test),"\n",'y_train개수: ', len(y_train),"y_test개수: ",len(y_test))
#train_test_split(x value, y value, test_size = ~~, random_state = ~)
#test_size = 0.2 => 20%만 사용한다 테스트에
#random_state = => 얼마나섞을것인지 난수 만들기
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

#model 학습부분
decision_tree = DecisionTreeClassifier(random_state=32) #모델 종류, 난수설정
decision_tree.fit(X_train, y_train) #모델 학습시키기(fitting)
y_pred = decision_tree.predict(X_test) #y_pred에 모델 예측값 삽입


accuracy = accuracy_score(y_test, y_pred) 
print("accuracy : ",accuracy) #정답데이터/전테데이터
print(classification_report(y_test, y_pred, target_names = cancer.target_names))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

#model 학습부분
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)


accuracy = accuracy_score(y_test, y_pred) 
print("accuracy : ",accuracy,"%") #정답데이터/전테데이터
print(classification_report(y_test, y_pred,target_names = cancer.target_names))

```

```python
#SVM
from sklearn import svm

#model 학습부분
svm_model = svm.SVC() #각각의 분류기마다 자기만의 식이 있다, 그걸 알아야함
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred) 
print("accuracy : ",accuracy) #정답데이터/전테데이터
print(classification_report(y_test, y_pred, target_names = cancer.target_names)) #타겟네임 추가함
```

```python
from sklearn.linear_model import SGDClassifier

#모델 학습부분
sgd_model = SGDClassifier() #모델 변수에 할당
sgd_model.fit(X_train, y_train) #모델에 데이터 할당
y_pred = sgd_model.predict(X_test) #정답치 변수 할당


accuracy = accuracy_score(y_test, y_pred) 
print("accuracy : ",accuracy)
print(classification_report(y_test, y_pred,target_names = cancer.target_names, zero_division = 0)) #정확도 메서드 이용해 정확도 판별
#확률적 경사하강 모델이라 할때마다 정확도가 변경된다
```

```python
#성능평가지표1 - 오차행렬 : 복원율(Recall값)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred) #지금은 바로 윗라인의 로지스틱 회귀모델로 산출한 결괏값의 신뢰되 파악

# 이진 표기법의 경우 오차행렬은
# TP FP
# FN TN
#이다. 해당 데이터는 암을 진단하는 알고리즘인데, 음성을 양성으로 판단하는 오류보다
#양성을 음성으로 판단하는 오류가 훨씬 더 위험하다. 따라서 FN값으로 평가하는 것이 중요한데,
#이를 나타내주는 지표가 복원율, RECALL이다. 복원율은 FN값이 클수록 감소하므로,
#복원율을 기준으로 알고리즘의 성능을 평가하는 것이 올바르다 추정된다.

#다만, 해당알고리즘은 암 확진이 0, 암 비확진이 1이므로 
#암 확진임에도 암 비확진으로 판별하는, 즉 0임에도 1로 판별하는, 
# 다시말해 Negative임에도 Positive로 판별하는 False Positive, FP의 값이 적을수록 좋은 알고리즘이다.

#해당 알고리즘의 FP는 11로, 암임에도 암이 아니라고 말한 경우가 11번
#암이 아님에도 암이라고 말한 경우가 1번(FN)
#암이기에 암이라고 말한경우가 115번
#암이 아니라 암이 아니라고 말한 경우가 44번이다.

#물론 이중 가장 중요한 평가척도는 FP, 암임에도 암이 아니라고 말한 경우의 수다.
```
