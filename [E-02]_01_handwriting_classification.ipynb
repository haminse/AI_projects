```python
#필요 모듈 import
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report


```

```python
digits = load_digits() #데이터 변수 할당
digits.keys()
digits_data = digits.data #feature data
digits_label = digits.target #target data
#digits_label.shape #label data를 feature data 맨 뒤에 붙일 예정임
digits.keys()
```

```python
digits.target_names #target names
```

```python
print(digits.DESCR) #target describe
```

```python
import pandas as pd #데이터 시각화 해서 확인먼저 하기

digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names)
digits_df
```

```python
#feature data에 label data 삽입
digits_df["label"] = digits_label #columns 이름 label
digits_df

#<error review>
#처음에 digits_data["label"]로 붙이려고 해서 type error가 뜸
#digits_data는 pd 형식의 자료가 아닌 단순 어레이구조기때문에
#해당 코드로 열 추가 불가능 pd형으로 바꾼 자료형에 열 추가 가능
```

```python
# (3) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state=7)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test),"\n",'y_train개수: ', len(y_train),"y_test개수: ",len(y_test))
#train_test_split(x value, y value, test_size = ~~, random_state = ~)
#test_size = 0.2 => 20%만 사용한다 테스트에
#random_state = => 얼마나섞을것인지 난수 만들기
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

#model 학습부분
decision_tree = DecisionTreeClassifier(random_state=32) #모델 종류, 난수설정
decision_tree.fit(X_train, y_train) #모델 학습시키기(fitting)
y_pred = decision_tree.predict(X_test) #y_pred에 모델 예측값 삽입


accuracy = accuracy_score(y_test, y_pred) #당연히 크기가 같아야겠쥬?
print("accuracy : ",accuracy) #정답데이터/전테데이터
print(classification_report(y_test, y_pred))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

#model 학습부분
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)


accuracy = accuracy_score(y_test, y_pred) #당연히 크기가 같아야겠쥬?
print("accuracy : ",accuracy) #정답데이터/전테데이터
print(classification_report(y_test, y_pred))
```

```python
#SVM
from sklearn import svm

#model 학습부분
svm_model = svm.SVC() #각각의 분류기마다 자기만의 식이 있다, 그걸 알아야함
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred) #당연히 크기가 같아야겠쥬?
print("accuracy : ",accuracy) #정답데이터/전테데이터
print(classification_report(y_test, y_pred))
```

```python
# SGD Classifier
from sklearn.linear_model import SGDClassifier

#모델 학습부분
sgd_model = SGDClassifier() #모델 변수에 할당
sgd_model.fit(X_train, y_train) #모델에 데이터 할당
y_pred = sgd_model.predict(X_test) #정답치 변수 할당


accuracy = accuracy_score(y_test, y_pred) #당연히 크기가 같아야겠쥬?
print("accuracy : ",accuracy)
print(classification_report(y_test, y_pred)) #정확도 메서드 이용해 정확도 판별
#확률적 경사하강 모델이라 할때마다 정확도가 변경된다
```
